```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

from shapely.geometry import Point, Polygon
import geopandas as gpd
```

## Finding a map

To be able to visualize the original Chernobyl data, I needed to find a map to place the data points on. 
I found this map (one cm is 500 km) on Natural Earth's website. 

```python


shapefile = 'ne_50m_admin_0_countries.shp'

# We set ISO_A2 since this is the country code style found in our Chernobyl dataset
gdf = gpd.read_file(shapefile)[['ADMIN', 'ISO_A2', 'geometry']]
gdf.columns = ['country', 'country_code', 'geometry']
gdf.head()
```

```python
gdf.plot()
```

## Matching up those countries

Now that I had a map, I needed to isolate the parts of the map related to the Chernobyl dataset. In other words, I needed to find the countries in Chernobyl and use them to select the right countries on the map. 

```python
air = pd.read_csv('processed_chernair.csv')
```

```python
air_countries = set(air.country)
map_countries = set(gdf.country_code)
```

```python
air_countries.difference(map_countries)
```

Using basic set theory, we can tell that F, NO, and UK are within the Chernobyl dataset, but they aren't within the map. 
Intuitively, we think F is just France, NO is Norway, and UK is the United Kingdom, but we need to be sure. 

```python
air.loc[air['country'] == 'F'].head()
```

A quick online search confirms that Cadarache is indeed a city in France and the longitude and latitude coordinates are correct. So, I can set F to be FR to follow ISO_A2 standards. 

```python
air.loc[air['country'] == 'F', 'country'] = 'FR'
```

Now, I look up France in the map datset. 

```python
gdf[gdf['country'].str.match('^F.*') == True] 
```

Interestingly enough, the country code is set to -99 here instead of FR. Let's replace it with FR. 

```python
gdf.loc[gdf['country'] == 'France', 'country_code'] = 'FR'
gdf.loc[gdf['country'] == 'France']
```

It turns out we have a similar situation with Norway. 

```python
gdf[gdf['country'].str.match('^N.*') == True]
```

```python
gdf.loc[gdf['country'] == 'Norway', 'country_code'] = 'NO'
gdf.loc[gdf['country'] == 'Norway']
```

Our last problematic entry was the UK.

```python
gdf[gdf['country'].str.match('^U.*') == True]
```

It turns out they use GB instead of UK. So I replace UK with GB in the Chernobyl dataset. 

```python
air.loc[air['country'] == 'UK', 'country'] = 'GB'
air.loc[air['country'] == 'GB']
```

If we redo the subsetting, we should get a complete match.

```python
air_countries = set(air.country)
map_countries = set(gdf.country_code)
```

```python
air_countries.difference(map_countries)
```

An empty set signifies no difference existed between the two sets. So they are equal. 

To recap, we found the countries represented in the Chernobyl dataset, and found these countries in the dataset. 
Then, we made sure their country codes matched. 

## Plotting the right countries
Now, we plot the countries from the Chernobyl dataset. 

```python
indices_chernobyl = []
for index in range(len(gdf)):
    if gdf.country_code.values[index] in air_countries:
        indices_chernobyl.append(index)
```

```python
gdf_map = gdf.iloc[indices_chernobyl,]
```

```python
gdf_map.index = range(len(gdf_map))
```

```python
gdf_map
```

```python
gdf_map.plot()
```

The plot seems to be stretched out given that Australia and other islands are on it. After plotting data, we'll be able to see which parts of the map we can eliminate to allow for better resolution. 

I suspect our map qualifies all territories belonging to a country as that country. For example, plotting France doesn't just plot continental europe France, but the island territories.

## Placing the data on the map
Now that we have a background map, we can try placing the concentration data on the map. Doing so merely requires changing our original dataset into a Geopandas Data Frame. 

```python
geo_air = gpd.GeoDataFrame(air)
```

```python
crs =  'epsg:4326'
geometry = [Point(xy) for xy in zip( air["longitude"], air["latitude"]  )] 
```

```python
geo_air = gpd.GeoDataFrame(air, crs = crs, geometry = geometry)
```

```python
geo_air.head()
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
gdf_map.plot(ax = ax, color = 'gray')
geo_air.plot(ax = ax, markersize = 5, color = 'r', alpha = 0.5)
```

Most of our markers fall where we'd expect them to (in Europe). Note, however, the pile of dots near (50, 20) and the solitary dot at (0,0). It's unlikely coordinates were taken in the ocean so we need to find out why these are plotted this way. 

```python
# Find points plotted below 22 degrees in latitude. 
index_counter = 0
index_list = []
for point in geo_air.geometry.values:
    if point.xy[1][0] < 22:
        index_list.append(index_counter)
    index_counter += 1
```

```python
geo_air.iloc[index_list].geometry
```

So, we found our points. Recall that the second coordinate of the point is the latitude. 

```python
low_points = geo_air.iloc[index_list]
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
gdf_map.plot(ax = ax, color = 'gray')
low_points.plot(ax = ax, markersize = 5, color = 'r', alpha = 0.5)
```

We can see we did indeed select the right points from the map. Now we try to find where we should place these points. 

```python
low_points['country'].groupby([low_points['city']]).describe()
```

Before continuing, we want to be sure about something. Is it safe to assume that these entries are comprehensive in the set? In other words, is it the case that Vaernes in the original dataset always mapped to these points (similarly for CZ and NO data)? An easy way to check is performing the same groupby and selection the cities we're interested in. 

```python
geo_air['country'].groupby([geo_air['city'] == 'VAERNES']).describe()
```

This matches the data about Vaernes in the low points so when we find out the issue, we will replace all Vaernes data in the original set. 

```python
geo_air['country'].groupby([geo_air['city'] == 'VALENCIA']).describe()
```

Similarly, we will be replacing all Valencia data in the original set. Finally, we check out the CZ part of our data. 

```python
geo_air['city'].groupby([geo_air['country'] == 'CZ']).describe()
```

```python
low_points['city'].groupby([low_points['country'] == 'CZ']).describe()
```

Note that the count of unique cities in CZ is 8 and there are 107 measurements. These match up for both sets so we will be replacing all CZ data in the original set. 

Most of our low points should be in CZ, but two points should be in NO and ES. Recall these are codes for Spain and Norway. An internet search reveals that Vaernes and Valencia should both be in continental europe. But, for some reason they both had (0,0) coordinates (see below). So, that singular point at (0,0) actually masked two difference city measurements.  

```python
low_points.loc[low_points['city'] == 'VAERNES']
```

```python
low_points.loc[low_points['city'] == 'VALENCIA']
```

We give these cities the appropriate lat and long measures. 

```python
low_points.loc[low_points.city == 'VAERNES', 'latitude'] = 59.199860
low_points.loc[low_points.city == 'VAERNES', 'longitude'] = 10.327740
low_points.loc[low_points.city == 'VAERNES'].head()
```

```python
low_points.loc[low_points.city == 'VALENCIA', 'longitude'] = -0.376288 
low_points.loc[low_points.city == 'VALENCIA', 'latitude'] = 39.469906
low_points.loc[low_points.city == 'VALENCIA'].head()
```

Now consider the CZ points. 

```python
low_points.loc[low_points.country == 'CZ']
```

You'll note these already have long and lats. An internet search reveals, for example, that CZ, Banska should have long 19.17 and lat 48.73 instead of what's in the data. In other words, the lat and long got switched! So, we switch the lats and longs back to what they should be for CZ. 

```python
old_lats = low_points.loc[low_points.country == 'CZ'].latitude
low_points.loc[low_points.country == 'CZ', 'latitude'] = low_points.loc[low_points.country == 'CZ'].longitude
low_points.loc[low_points.country == 'CZ', 'longitude'] = old_lats
low_points.loc[low_points.country == 'CZ']
```

Finally, we modified the long and lat colums, but we still need to modify the geometry column. This column interacts with the Geopandas package for plotting. 

```python
# Redo the point zipping 
geometry = [Point(xy) for xy in zip( low_points["longitude"], low_points["latitude"]  )] 
low_points.geometry = geometry
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
gdf_map.plot(ax = ax, color = 'gray')
low_points.plot(ax = ax, markersize = 5, color = 'r', alpha = 0.5)
```

Now, our plot makes more sense. 

We now update our original air data to include these points with their proper lats and longs. 

```python
geo_air.loc[geo_air.country == 'CZ'] = low_points.loc[low_points.country == 'CZ']
geo_air.loc[geo_air.city == 'VAERNES'] = low_points.loc[low_points.city == 'VAERNES']
geo_air.loc[geo_air.city == 'VALENCIA'] = low_points.loc[low_points.city == 'VALENCIA']
```

Our plot should now have all points in continental europe. 

```python
fig, ax = plt.subplots(figsize=(12, 8))
gdf_map.plot(ax = ax, color = 'gray')
geo_air.plot(ax = ax, markersize = 2, color = 'r', alpha = 0.5)
```

## Australia and friends

(4/6/2020 A simpler solution is just modifying axes. But at least we practice removing countries in these data below)
The below is useful and important to showing appropriate data since Iran and Australia shouldn't have been in the plot. 

Looking at the plot above, we can tell we need to remove places on the map without any Chernobyl data. Before removing places though, we need to be sure these places don't have data in the Chernobyl set, and we just didn't map it properly. 

```python
'AU' in geo_air.country.values # Australia
```

```python
geo_air.loc[geo_air.country == 'AU']
```

Here's the issue: the country code here suggests its Australia, but in reality this is Austria (Vienna gave it away). In two letter code, this should be AT. Let's make the switch.  

```python
geo_air.loc[geo_air.country == 'AU', 'country'] = 'AT'
```

Now, we need to remove the Australia from map data and add in Austria. 

```python
gdf_map
```

```python
au_index = gdf_map.index[gdf_map.country == 'Australia']
au_index = au_index[0]
```

```python
gdf_map.at[au_index, 'country'] = 'Austria'
gdf_map.at[au_index, 'country_code'] = 'AT'
gdf_map.at[au_index, 'geometry'] = gdf.loc[gdf.country_code == 'AT', 'geometry'].values[0]
```

Now our plot should not have Australia in it. 

```python
fig, ax = plt.subplots(figsize=(12, 8))
gdf_map.plot(ax = ax, color = 'gray')
geo_air.plot(ax = ax, markersize = 2, color = 'r', alpha = 0.5)
```

Continuing our plot desctrution scheme, notice Iran's shape on the far right side. Did our Chernobyl data really have data from Iran?

```python
geo_air.loc[geo_air.country == 'IR'].head()
```

The data set did not have data frome Iran. By current standards, Ireland's country code is actually IE, not IR. Like we did for Australia, we'll remove Iran. But first, we change IR to IE in the original data. 

```python
geo_air.loc[geo_air.country == 'IR', 'country'] = 'IE'
```

Now we replace Iran with Ireland in the map data.

```python
ir_index = gdf_map.index[gdf_map.country == 'Iran']
ir_index = ir_index[0]
```

```python
gdf_map = gdf_map.drop(ir_index)
```

```python
gdf_map = gdf_map.append(gdf.loc[gdf.country == 'Ireland'])
```

```python
gdf_map.index = range(len(gdf_map))
```

```python
gdf_map
```

We can now view the map without Iran, but with Ireland included. 

```python
fig, ax = plt.subplots(figsize=(12, 8))
gdf_map.plot(ax = ax, color = 'gray')
geo_air.plot(ax = ax, markersize = 2, color = 'r', alpha = 0.5)
```
